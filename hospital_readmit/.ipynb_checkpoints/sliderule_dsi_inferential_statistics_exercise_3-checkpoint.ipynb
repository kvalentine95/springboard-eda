# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

<h4>Compute the observed significance value</h4>
<p>In our case, it would be useful to determine the Pearson correlation coefficient between the number of discharges and the excess readmission rate as it would measure the strength of a linear association between two variables. The value of the Pearson correlation ranges from -1 to 1 with 0 denoting the absence of correlation between two variables. According to the preliminary scatter plot, we could expect that the calculated Pearson correlation coefficient is negative as the high excess readmission rate is negatively correlated with the number of discharges.</p>

```python
# check for any NaN values
clean_hospital_read_df.isnull().sum()
```

```python
# eliminate the rows with no info about the excess readmission ratio
clean_hospital_read_df = clean_hospital_read_df.dropna(subset=['Excess Readmission Ratio'])
clean_hospital_read_df.info()
```

```python
# define the pearson correlation coefficient
def pearson_r(x,y):
    corr_mat = np.corrcoef(x,y)
    return corr_mat[0,1]
num_discharge = clean_hospital_read_df['Number of Discharges']
excess_ratio = clean_hospital_read_df['Excess Readmission Ratio']
r = pearson_r(excess_ratio, num_discharge)
print('The Pearson correlation coefficient is ' + str(r))
```

It seems like our observed Pearson correlation coefficient shows that there is indeed a negative correlation between the excess readmission ratio and the number of discharges. However, the observed correlation may just be by chance. That is why we need to perform permutation.

```python
# create permutation replicates and permute excess_ratio array while keeping the num_discharge constant
perm_replicates = np.empty(10000)
for i in range(10000):
    excess_permuted = np.random.permutation(excess_ratio)
    perm_replicates[i] = pearson_r(excess_permuted,num_discharge)
p = np.sum(perm_replicates <= r)/len(perm_replicates)
print('p-val = ' + str(p))
```

<h4>Report statistical significance for  α  = .01.</h4>

The p-value is small enough (much smaller than alpha) for us to reject the null hypothesis. Therefore, we can conclude that the correlation between the excess ratio of hospital readmissions and the number of discharges is statistically significant.

<h4>Construct another plot that conveys the same information in a more direct manner.</h4>

The plot provided does not show any correlations and contains arbitrary boxes for small and large discharge values. The advantage of this plot is that it shows an entire population and demonstrates an overall trend in the relationship. We can perform a linear regression using np.polyfit() to show how the excess ratio of hospital readmissions varies with the number of discharges. 

```python
# plot the excess ratio versus number of discharges
_=plt.plot(num_discharge,excess_ratio, marker='.',linestyle='none',alpha=0.2)
plt.margins(0.02)
_=plt.xlabel('Number of Discharges')
_=plt.ylabel('Excess Ratio of Readmissions')
_=plt.xlim((0,3050))
# perform a linear regression using np.polyfit()
a,b = np.polyfit(num_discharge, excess_ratio, deg=1)
# make theoretical line to plot
x = np.array([0,3000])
y = a*x + b
# add regression line to the plot
_ = plt.plot(x,y,color='magenta')
plt.show()
```
